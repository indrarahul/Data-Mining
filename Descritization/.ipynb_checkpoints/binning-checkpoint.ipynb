```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('../Clustering/iris.csv')
```

```python
datamodify = np.array(data)
datamodify1 = np.array(data)
datamodify2 = np.array(data)
data = np.array(data)
```

```python
bin_size = 15
data_size = 150
```

```python
def bin_by_mean():
  mean = 0
  for i in range(0,data_size,bin_size):
    for j in range(bin_size):
      mean+=datamodify[i+j][0]
    mean/=bin_size
    for j in range(bin_size):
      datamodify[i+j][0]=round(mean,1)
```

```python
def bin_by_median():
  median = []
  for i in range(0,data_size,bin_size):
    median = []
    for j in range(bin_size):
      median.append(datamodify1[i+j][0])
    median.sort()
    median = np.array(median)
    for j in range(bin_size):
      datamodify1[i+j][0]=round(np.median(median),1)
```

```python
def bin_by_boundry():
  mean = 0
  for i in range(0,data_size,bin_size):
    for j in range(1,bin_size-1):
      if(abs(datamodify2[i+j][0]-datamodify2[i][0])>abs(datamodify2[i+j][0]-datamodify2[i+(bin_size-1)][0])):
        datamodify2[i+j][0] = datamodify2[i+(bin_size-1)][0]
      else:
        datamodify2[i+j][0] = datamodify2[i][0]
```

```python
def print_dif(a,b):
  print("\tORIGINAL\t\tNEW\t\t\tERROR")
  for i in range(150):
    print(str(i+1) +"\t"+ str(a[i][0])+'\t\t\t'+str(b[i][0]) + '\t\t\t' + str(round(abs(a[i][0]-b[i][0]),1)))
```

# **Binning By Mean Method**

```python
bin_by_mean()
```

```python
print_dif(data,datamodify)
```

# **Binning By Median Method**

```python
bin_by_median()
```

```python
print_dif(data,datamodify1)
```

# **Binning By Boundry Method**

```python
bin_by_boundry()
```

```python
print_dif(data,datamodify2)
```
